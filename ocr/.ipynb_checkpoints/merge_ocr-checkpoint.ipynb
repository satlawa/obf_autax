```python
import os
import numpy as np
import pandas as pd
```

```python
# set home directory
home_path = os.getcwd()[:os.getcwd().find('Code/python/obf_autax/ocr')-1]
# set input_path
sap_path = os.path.join(home_path, 'Data', 'obf', 'roh_171_12.XLS')
ma_path = os.path.join(home_path, 'Data', 'obf', 'nutzungen_171_12.csv')
```

```python
ma_data = pd.read_csv(ma_path, sep=',')
sap_data = pd.read_csv(sap_path, sep='\t', encoding = "ISO-8859-1", decimal=',', error_bad_lines=False)
```

```python
ma_data
```

```python
sap_data
```

```python
sap_data = sap_data[sap_data['Best.-Schicht'] == 0].copy()
```

```python
sap_data.sort_values(by=['Abteilung', 'WE-Typ', 'Unterabteil.', 'Teilfl.'], ascending=[True,False,True,True], inplace=True)
sap_data = sap_data[['Waldort', 'WE-Typ', 'Umtriebszeit', 'Ertragssituation',
       'Bewirtschaftungsform', 'Schutzwaldkategorie', 'Fläche in HA',
       'Seehöhe', 'Exposition', 'Neigung', 'Standorteinheit', 'Alter der 1. Schicht']]
```

```python
ma_data.columns = ['S', 'MA', 'FL', 'LH', 'NH', 'SUM', 'DR', 'BH', 'ZP', 'SG', 'RU', 'Text', 'Waldort']
```

```python
ma_data
```

```python
merge_data = pd.merge(sap_data, ma_data, how='left', on='Waldort')
```

```python
merge_data
```

```python
data_path = os.path.join(home_path, 'Data', 'obf')
merge_data.to_csv(data_path + "/nutz_171_12.csv", index=False)
```
