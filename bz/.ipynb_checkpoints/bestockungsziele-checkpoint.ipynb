```python
import os
import numpy as np
import pandas as pd
import re
```

```python
# set home directory
home_path = os.getcwd()[:os.getcwd().find('Code/python/obf_autax/bz')-1]
# set bz_path
bz_path = os.path.join(home_path, 'data', 'obf', 'merge', 'bz', 'TO_1356_01_20200113_bz.csv')

print(home_path)
print(bz_path)
```

```python
bz_data = pd.read_csv(bz_path, sep=',')
```

```python
bz_data.head()
```

```python
# filter just WO
bz_data = bz_data.loc[bz_data['WE-Typ']=='WO', :]
```

```python
bz_data = bz_data[['WO', 'BZ1', 'BZ2', 'BZ3', 'BZ4']]
```

```python
bz_data.iloc[0,:]
```

```python
bz = bz_data.copy()
```

```python
bz = bz.fillna(0)
```

### create empty dataframe

```python
index = bz['WO'].unique()
```

```python
columns = ['FI', 'TA', 'LA', 'KI', 'DG', 'TH', 'BU', 'EI', 'HB', 'AH', 'ES', 'ER', 'RE', 'SP', 'EL', 'SN', 'SL']
```

```python
df = pd.DataFrame(index=index, columns=columns)
df = df.fillna(0) # with 0s rather than NaNs
```

```python
df
```

### change structure

```python
for i in range(bz.shape[0]):
    row = bz.iloc[i,:]
    wo = row['WO']
    for j in range(1, bz.shape[1]):
        if row.iloc[j] != 0:
            bz_x = re.findall(r"[\w']+", row.iloc[j])
            df.loc[wo,bz_x[1]] = bz_x[0]
```

```python
df['Waldort'] = df.index
```

```python
df
```

```python
data_path = os.path.join(home_path, 'data', 'obf', 'merge', 'bz')
df.to_csv(data_path + "/bz_171_01.csv", index=False)
```
