```python
import os
import numpy as np
import pandas as pd
import re
```

```python
# set home directory
home_path = os.getcwd()[:os.getcwd().find('Code/python/obf_autax/bz')-1]
# set bz_path
bz_path = os.path.join(home_path, 'Data', 'obf_autman', 'sources', 'bz_171_12_ps.xlsx')

print(home_path)
print(bz_path)
```

```python
bz_data = pd.read_excel(bz_path)
```

```python
bz_data.head()
```

```python
# filter just WO
bz_data = bz_data.loc[bz_data['WE-Typ']=='WO', :]
```

```python
bz_data = bz_data[['WO', 'BZ1', 'BZ2', 'BZ3', 'BZ4', 'BZ5']]
```

```python
bz_data.iloc[0,:]
```

```python
bz = bz_data.copy()
```

```python
bz = bz.fillna(0)
```

### create empty dataframe

```python
index = bz['WO'].unique()
```

```python
columns = ['FI', 'TA', 'LA', 'KI', 'SK', 'DG', 'TH', 'BU', 'EI', 'HB', 'AH', 'ES', 'ER', 'RE', 'LI', 'SP', 'EL', 'SN', 'SL']
```

```python
df = pd.DataFrame(index=index, columns=columns)
df = df.fillna(0) # with 0s rather than NaNs
```

```python
df
```

### change structure

```python
for i in range(bz.shape[0]):
    row = bz.iloc[i,:]
    wo = row['WO']
    for j in range(1, bz.shape[1]):
        if row.iloc[j] != 0:
            bz_x = re.findall(r"[\w']+", row.iloc[j])
            df.loc[wo,bz_x[1]] = bz_x[0]
```

```python
df['Waldort'] = df.index
```

```python
df
```

```python
# set raw_path
raw_path = os.path.join(home_path, 'Data', 'obf_autman', 'sources', '1310_171_12_20200821.XLS')
# read tsv
wo_data = pd.read_csv(raw_path, sep='\t', encoding = "ISO-8859-1", decimal=',', error_bad_lines=False)
# filter relevant data
wo_data = wo_data[wo_data['Best.-Schicht'] == 0]

### pernitz
wo_data = wo_data[wo_data['Abteilung'] > 200]
```

```python
# merge

wo_data.sort_values(by=['Abteilung', 'WE-Typ', 'Unterabteil.', 'Teilfl.'], ascending=[True,False,True,True], inplace=True)
wo_data = wo_data[['Waldort', 'WE-Typ']]

merge_data = pd.merge(wo_data, df, how='left', on='Waldort')
```

```python
merge_data.iloc[:,2:].astype(int)
```

```python
merge_data.fillna(0, inplace = True)
merge_data.iloc[:,2:] = merge_data.iloc[:,2:].astype(int)
```

```python
# set bz_path
save_path = os.path.join(home_path, 'Data', 'obf_autman', 'csv', 'bz_autax_1310_171_12_20200821.XLS')
merge_data.to_csv(save_path, index=False)
```
