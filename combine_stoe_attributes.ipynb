```python
import os
import numpy as np
import pandas as pd
import re
```

```python
# set home directory
home_path = os.getcwd()[:os.getcwd().find('Code/python/obf_autax')-1]
# set bz_path
new_path = os.path.join(home_path, 'Data', 'obf_autman', 'csv', 'johannklause', 'xwo_data_1366_08.csv')
raw_path = os.path.join(home_path, 'Data', 'obf_autman', 'raw_data', 'TO_1366_181_08_20201111.XLS')

print(home_path)
print(new_path)
```

```python
# read csv
data_new = pd.read_csv(new_path)
# read tsv
wo_data = pd.read_csv(raw_path, sep='\t', encoding = "ISO-8859-1", decimal=',', error_bad_lines=False)
# filter relevant data
wo_data = wo_data[wo_data['Best.-Schicht'] == 0]

# sort data
wo_data.sort_values(by=['Abteilung', 'WE-Typ', 'Unterabteil.', 'Teilfl.'], ascending=[True,False,True,True], inplace=True)
```

```python
data_new = data_new.fillna(0)
```

```python
data_new
```

```python
wo_data = wo_data.reset_index(drop=True)
wo_data = wo_data[['Waldort', 'WE-Typ', 'Betriebsklasse', 'Umtriebszeit', 'Standorteinheit', 'Vegetationstyp', 'Verbissgrad']]
```

```python
wo_data
```

```python
print(data_new.shape)
print(wo_data.shape)
```

```python
for i in range(0,1790):
    if data_new.loc[i,'bkl'] != 0:
        wo_data.loc[i,'Betriebsklasse'] = data_new.loc[i,'bkl']
    if data_new.loc[i,'uz'] != 0:
        wo_data.loc[i,'Umtriebszeit'] = data_new.loc[i,'uz']
    if data_new.loc[i,'stoe'] != 0:
        wo_data.loc[i,'Standorteinheit'] = data_new.loc[i,'stoe']
    if data_new.loc[i,'vtyp'] != 0:
        wo_data.loc[i,'Vegetationstyp'] = data_new.loc[i,'vtyp']
    if data_new.loc[i,'vg'] != 0:
        wo_data.loc[i,'Verbissgrad'] = data_new.loc[i,'vg']
```
