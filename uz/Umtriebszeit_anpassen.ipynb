```python
import numpy as np
from numpy import genfromtxt
import pandas as pd
```

```python
# load data
data = pd.read_excel('/Users/philipp/Python/bonitaet_fe/to_1345_06.xlsx')
```

```python
data
```

```python
try:
    data['WO'] = data['Abteilung'].apply(str) + data['Unterabteil.'] + data['Teilfl.'].apply(str)
except TypeError:
    data['WO'] = data['Abteilung'].apply(str) + data['Unterabteil.'].apply(str) + data['Teilfl.'].apply(str)
```

```python
data_wo = data[data['Best.-Schicht']==0]
```

```python
data_wo = data_wo.fillna(0)
```

```python
data_wo = data_wo[['WO', 'Abteilung', 'Unterabteil.', 'Teilfl.', 'WE-Typ', 'Umtriebszeit', 'Nebengrund Art', 'Ertragssituation', 'Bewirtschaftungsform', 'Schutzwaldkategorie', 'Fläche in HA']]
```

```python
data_wo = data_wo.sort_values(by=['Abteilung', 'Nebengrund Art', 'Unterabteil.', 'Teilfl.'])
```

```python
data_wo.to_excel('/Users/philipp/Python/bonitaet_fe/xtest.xlsx', index=False)
```

```python
data_x = data[data['Baumart']=='FI']
```

```python
data_x = data_x[['WO', 'Ertragsklasse']]
```

```python
data_x
```

```python
data_merge = pd.merge(data_wo, data_x, how='left', on='WO')
```

```python
data_merge
```

```python
data_merge['UZ'] = data_merge['Umtriebszeit']
```

```python
data_merge.loc[:,'UZ']
```

```python
# do not use it like this
data_merge[(data_merge['Bewirtschaftungsform']=='S') & (data_merge['Ertragssituation']=='A')]['UZ'] = 200
```

```python
# use .loc instead
data_merge.loc[((data_merge['Bewirtschaftungsform']=='S') & (data_merge['Ertragssituation']=='A')),'UZ'] = 200
data_merge.loc[((data_merge['Bewirtschaftungsform']=='S') & (data_merge['Ertragssituation']=='I')),'UZ'] = 160
data_merge.loc[((data_merge['Bewirtschaftungsform']=='W') & (data_merge['Ertragssituation']=='A')),'UZ'] = 160
data_merge.loc[((data_merge['Bewirtschaftungsform']=='W') & (data_merge['Ertragssituation']=='I') & 
                (data_merge['Ertragsklasse'] <= 7)),'UZ'] = 140
data_merge.loc[((data_merge['Bewirtschaftungsform']=='W') & (data_merge['Ertragssituation']=='I') & 
                (data_merge['Ertragsklasse'] > 7)),'UZ'] = 120
```

```python
data_merge = data_merge.drop_duplicates('WO')
```
